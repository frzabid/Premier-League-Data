```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# URL untuk mengambil data
url = "https://www.bing.com/sportsdetails?q=fixtures%20epl&sport=Soccer&scenario=League&TimezoneId=SE%20Asia%20Standard%20Time&league=Soccer_EnglandPremierLeague&intent=Schedule&seasonyear=2024&segment=sports&isl2=true&"

# Mengambil konten dari halaman
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Mencari elemen dengan kelas "fixtures__date-container"
fixtures_container = soup.find_all("div", class_="bsp-l2-card-grp bsp_schedule_mtch_crd soccer")

fixtures = []

# Mengecek apakah elemen ditemukan
if fixtures_container:
    for container in fixtures_container:
        # Mendapatkan teks di dalam elemen tersebut
        content_text = container.get_text(strip=True)
        fixtures.append(content_text)

    # Menggabungkan semua teks menjadi satu string
    all_fixtures = ' '.join(fixtures)

    # Menggunakan regex untuk mengekstrak informasi
    pattern = r'(?P<date>\w{3}, \w{3} \d{1,2})·FTWeek \d(?P<home_team>[A-Za-z\s]+)(?P<home_score>\d)(?P<away_team>[A-Za-z\s]+)(?P<away_score>\d)'
    matches = re.finditer(pattern, all_fixtures)

    # Menyusun hasil ke dalam list
    structured_fixtures = []
    for match in matches:
        structured_fixtures.append({
            'Tanggal': match.group('date'),
            'Tim Tuan Rumah': match.group('home_team').strip(),
            'Skor': f"{match.group('home_score')} - {match.group('away_score')}",
            'Tim Tamu': match.group('away_team').strip()
        })

    # Mengubah list ke dalam DataFrame
    df = pd.DataFrame(structured_fixtures)
    print(df)

    # Menyimpan DataFrame ke file CSV tanpa menyertakan indeks
    df.to_csv('fixtures_premier_league.csv', index=False)
else:
    print("Konten tidak ditemukan.")
```
